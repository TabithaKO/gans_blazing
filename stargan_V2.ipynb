- To test on custom source images go to the "dataset" directory and add a directory called "4_domains".

- In 4_domains add a "test" and "train" directory. 
- In "test" add "src_imgs" and "ref_imgs". At this point you can add source images to the "src_imgs" directory
- In the "ref_imgs" add "black_women", "black_men", "white_women", "white_men" and add the respective source images from these domains

NOTE 📝 : You get better results when you have portrait images with the person facing forward
